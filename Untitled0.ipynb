```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('Walmart.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
## checking missing values
total_missing_value = df.isnull().sum().sum()
missing_value_column = df.isnull().sum()

print(f'DATASET HAS TOTAL:\n{total_missing_value} missing values\n')
print(f'DATASET CONTAINS MISSING VALUE BY COLUMNS\n{missing_value_column}')
```

```python
## convert 'unit_price' to string, remove '$', and convert to float
df['unit_price'] = pd.to_numeric(df['unit_price'].astype(str).str.replace('$', '', regex=False), errors='coerce')
df['unit_price'] = df['unit_price'].astype(float)
## calculate the mean
mean_unit_price = df['unit_price'].mean()
mean_quantity = df['quantity'].mean()
## replace NaN values with the calculated mean for
df['unit_price'] = df['unit_price'].fillna(mean_unit_price)
df['quantity'] = df['quantity'].fillna(mean_quantity)

print(f"total_missing_value: {df.isnull().sum().sum()}")
```

```python
## check for duplicate values
duplicates = df.duplicated()
total_duplicate = duplicates.sum()
print(f'Total duplicate record:\n{total_duplicate}\n')
```

```python
## drop dublicates
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
## save cleaned dataset
df.to_csv('walmart_clean_data.csv', index=False)
```

```python
print("\nData Types:\n", df.dtypes)
# attribute types
nominal_attributes = ['invoice_id', 'Branch', 'City', 'category', 'payment_method','date','time']  # Examples
ordinal_attributes = ['rating']

print("\nNominal Attributes:", nominal_attributes)
print("Ordinal Attributes:", ordinal_attributes)
```

```python
## applied statistics
numerical_df = df.select_dtypes(include=np.number)
print("\nBasic Statistics:\n", numerical_df.describe())

if 'unit_price' in numerical_df.columns:
    unit_prices = numerical_df['unit_price'].values  ## convert to NumPy array
    print("\nUnit Price Statistics (NumPy):")
    print("Mean:", np.mean(unit_prices))
    print("Median:", np.median(unit_prices))
    print("Standard Deviation:", np.std(unit_prices))
    print("Variance:", np.var(unit_prices))
    print("Min:", np.min(unit_prices))
    print("Max:", np.max(unit_prices))
    print("Sum:", np.sum(unit_prices))
    print("Product:", np.prod(unit_prices))
```

```python
# find the average, minimum, and maximum rating of categories for each city
category_ratings = df.groupby(['City', 'category']).agg(
    min_rating=('rating', 'min'),
    max_rating=('rating', 'max'),
    avg_rating=('rating', 'mean')
).reset_index()

print(category_ratings)
```

```python
city_avg_ratings = category_ratings.groupby('City')['avg_rating'].mean()

## top 5 and bottom 5 cities based on average rating
top_5_cities = city_avg_ratings.nlargest(5).index
bottom_5_cities = city_avg_ratings.nsmallest(5).index

# Filter
top_5_data = category_ratings[category_ratings['City'].isin(top_5_cities)]
bottom_5_data = category_ratings[category_ratings['City'].isin(bottom_5_cities)]
combined_data = pd.concat([top_5_data, bottom_5_data])
combined_data['City_Group'] = combined_data['City'].apply(lambda x: 'Top 5' if x in top_5_cities else 'Bottom 5')

## Grouped Bar Chart with FacetGrid

g = sns.FacetGrid(combined_data, col="City", col_wrap=5, height=3, aspect=1, col_order=list(top_5_cities) + list(bottom_5_cities))
g.map(sns.barplot, "category", "avg_rating", order=combined_data.sort_values('avg_rating')['category'].unique())
g.map(sns.pointplot, "category", "min_rating", color='black', order=combined_data.sort_values('avg_rating')['category'].unique())
g.map(sns.pointplot, "category", "max_rating", color='black', order=combined_data.sort_values('avg_rating')['category'].unique())
g.set_xticklabels(rotation=45, ha="right", fontsize=8)
g.set_axis_labels("Category", "Rating")
g.set_titles("{col_name}")
plt.tight_layout()
plt.show()
```

```python
## combined Point Plot with Error Bars
plt.figure(figsize=(16, 8))
sns.pointplot(x="category", y="avg_rating", hue="City", data=combined_data, capsize=.2, errwidth=1, hue_order=list(top_5_cities) + list(bottom_5_cities))
plt.xticks(rotation=45, ha="right")
plt.xlabel("Category")
plt.ylabel("Average Rating")
plt.title("Average Rating with Min/Max Range (Top 5 and Bottom 5 Cities)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
payment_counts = df.groupby('payment_method').size().to_frame(name='no_payments').reset_index()

# 1. Bar Chart
plt.figure(figsize=(8, 6))
sns.barplot(x='payment_method', y='no_payments', data=payment_counts)
plt.title('Number of Payments by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Number of Payments')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
qty_sold_by_payment = df.groupby('payment_method')['quantity'].sum().reset_index()
qty_sold_by_payment = qty_sold_by_payment.rename(columns={'quantity': 'no_qty_sold'})

styled_df = qty_sold_by_payment.style.background_gradient(cmap='viridis')
display(styled_df)

plt.figure(figsize=(8, 8))
plt.pie(qty_sold_by_payment['no_qty_sold'], labels=qty_sold_by_payment['payment_method'], autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Quantity Sold by Payment Method')
plt.tight_layout()
plt.show()
```

```python
df['total_profit'] = df['unit_price'] * df['quantity'] * df['profit_margin']

profit_by_category = df.groupby('category')['total_profit'].sum().reset_index()
profit_by_category = profit_by_category.sort_values('total_profit', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='total_profit', y='category', data=profit_by_category)
plt.title('Total Profit by Category')
plt.xlabel('Total Profit')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

```python
## top 10 and bottom 10 cities based on average rating
top_10_cities = city_avg_ratings.nlargest(10).index
bottom_10_cities = city_avg_ratings.nsmallest(10).index

# Filter
top_10_df = df[df['City'].isin(top_10_cities)]
bottom_10_df = df[df['City'].isin(bottom_10_cities)]
# Calculate payment method counts for top and bottom cities
top_payment_counts = top_10_df['payment_method'].value_counts()
bottom_payment_counts = bottom_10_df['payment_method'].value_counts()

## Pie Charts
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

axes[0].pie(top_payment_counts, labels=top_payment_counts.index, autopct='%1.1f%%', startangle=90, textprops={'fontsize': 10}) #Adjust fontsize if needed
axes[0].set_title('Payment Method Distribution in Top 10 Cities', fontsize=12)

axes[1].pie(bottom_payment_counts, labels=bottom_payment_counts.index, autopct='%1.1f%%', startangle=90, textprops={'fontsize': 10}) #Adjust fontsize if needed
axes[1].set_title('Payment Method Distribution in Bottom 10 Cities', fontsize=12)

plt.tight_layout()
plt.show()
```
